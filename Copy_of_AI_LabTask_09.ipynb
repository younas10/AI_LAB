**Imports**

```python
from sympy import symbols
from sympy.logic.boolalg import And, Not, Implies
```

**Variables**

```python
x, y = symbols('x y')
```

**Boolean predicates**

```python
# Boolean symbols (not Function)
Bird, CanFly = symbols('Bird CanFly')
Cat, Animal = symbols('Cat Animal')
Person, Parent = symbols('Person Parent')
Raining, Wet = symbols('Raining Wet')
```

**TASK 1 (Simplified & Working)**

**Every cat is an animal**

```python
statement1 = Implies(Cat, Animal)
statement1
```

**If it is raining, then ground is wet**

```python
statement2 = Implies(Raining, Wet)
statement2
```

. **"For all x, if x is a person, then there exists a y such that y is the parent of x."**

```python
statement3 = Implies(Person, Parent)
statement3
```

**TASK 2 — Negation**

**Original: “Some birds can fly”**

```python
original_statement = And(Bird, CanFly)
original_statement
```

**Negation**

```python
negated_statement = Not(original_statement)
negated_statement
```
